```python
import pandas as pd
import plotly.express as px

file_path = '/content/drive/MyDrive/Portfolio Projects/HR_Analytics/HR_Analytics.csv' #define correct path

df=pd.read_csv(file_path) #load the csv file into pandas dataframe (df)
#inspect the loaded data
df.info()
df.head()
```

```python
df.drop(columns=['EmployeeCount', 'EmployeeNumber', 'Over18'], inplace=True) #drop unnecessary columns
df.head()
```

```python
#Break into generational groupings
def categorize_generation(age):
    if age < 25:
        return "Gen Z"
    elif 25 <= age < 40:
        return "Millennial"
    elif 40 <= age < 55:
        return "Gen X"
    else:
        return "Baby Boomer"

df['Generation'] = df['Age'].apply(categorize_generation)

df[['Generation', 'Age']].head()
```

```python
print(df['Generation'].value_counts()) #count how many of each generation exist
```

```python
import numpy as np #To get min and max age range of employees

youngest= np.min(df['Age'])
oldest= np.max(df['Age'])

print(f"Youngest_Employee_Age: {youngest}")
print(f"Oldest_Employee_Age: {oldest}")
```

```python
#Check min and max duration for employees for columns 'YearsAtCompany', 'YearsInCurrentRole', 'YearsWithCurrManager'

columns_to_check = ['YearsAtCompany', 'YearsInCurrentRole', 'YearsWithCurrManager']

for col in columns_to_check:
    min_value = np.min(df[col])
    max_value = np.max(df[col])
    print(f"{col}: Min = {min_value}, Max = {max_value}")
```

```python
for col in columns_to_check:
    min_value = np.min(df[col])
    max_value = np.max(df[col])

    print(f"\nEmployees with minimum {col}:")
    print(df[df[col] == min_value])

    print(f"\nEmployees with maximum {col}:")
    print(df[df[col] == max_value])
```

```python
#Sort top and bottom 5 to show range

for col in columns_to_check:
    print(f"\nBottom 5 employees for {col}:")
    print(df.nsmallest(5, col))  # Smallest 5 values

    print(f"\nTop 5 employees for {col}:")
    print(df.nlargest(5, col))  # Largest 5 values
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#compare attrition rates by generation - bar chart
plt.figure(figsize=(8,5))
sns.countplot(x='Generation', hue='Attrition', data=df, palette='viridis')
plt.title("Attrition_by_Generation")
plt.show()
```

To compare attrition across generations, will look at the attrition rate %.


```python
#checking attrition rates percentage

attrition_rates = df.groupby('Generation')['Attrition'].value_counts(normalize=True).unstack() * 100
print(attrition_rates)
```

```python
#attrition rate % by stacked bar chart

attrition_rates.plot(kind='bar', stacked=True, figsize=(8,5), colormap='viridis')
plt.title("Attrition_Rate_by_Generation (%)")
plt.ylabel ("Percentage")
plt.xticks(rotation=0)
plt.show()
```


* Gen Z are the highest-risk group for attrition with almost 4 in 10 (~39%)  Gen Z employees leaving the company.

* Millennials have the second highest attrition rate, although more employees. Although more stable than Gen Z, still a concern at ~16.7%.

* Gen X and Baby Boomers have the lowest attrition, with older employees tending to have more job stability, making them more likely to stay.

```python
#For more clarity to see trend by gender and generation using Line Chart for Satisfaction Scores

plt.figure(figsize=(10, 6))

#Calculate mean satisfaction scores grouped by gender and generation
satisfaction_means = df.groupby(['Generation', 'Gender'])[['JobSatisfaction', 'EnvironmentSatisfaction', 'RelationshipSatisfaction']].mean().reset_index()

#Convert from wide to long format for easier plotting
satisfaction_melted = satisfaction_means.melt(id_vars=['Generation', 'Gender'], var_name='Satisfaction Type', value_name='Score')

#Plot the lines
sns.lineplot(x='Generation', y='Score', hue='Gender', style='Satisfaction Type', markers=True, data=satisfaction_melted)

plt.title("SatisfactionScores_by_Gender&Generation")
plt.xlabel("Generation")
plt.ylabel("Average Satisfaction Score")
plt.legend(title="Satisfaction Type")
plt.show()
```

KEY TAKEAWAYS BY GENERATION
* **Baby Boomers** women prioritize relationship satisfaction, men prefer job satisfaction. Environmental satisfaction is lowest for both, but worse for men.
* **Gen X** both men and women score job and relationship satisfaction the same. However, men score environmental satisfaction higher.
* **Gen Z** women show the lowest score across ALL satisfaction categories, especially with the biggest gap in Environment & relationship satisfaction, suggesting Gen Z women may feel disconnected or unsupported. Workplace culture, inclusion, and career growth opportunities might be a major concern for Gen Z women.
* **Millennial** women are less satisfied across all areas and therefore might face similar challenges as Gen Z women but to a lesser extent.

**CONCLUSION** overall men generally report higher satisfaction scores across all generations, with Gen Z women least satisfied, with Millennial women following a similar pattern. Gen X women are the most balanced and Boomer women value relationships, while Boomer men focus on job satisfaction.

To finish sum up the attrition and trend, will look into ML to forecast future attrition trends.
* Will calculate attrition rates for each generation.
* Use Linear Regression to model attrition trends over time.
* Predict attrition for the next 10-30 years and visualise trends as Gen X naturally retire and the impact of Millennials and Gen Z leaving early, creating a potential workforce gap.

```python
#Attrition_Rates

attrition_rates = df.groupby('Generation')['Attrition'].value_counts(normalize=True).unstack() * 100
attrition_rates

#Rename columns for clarity
attrition_rates.columns = ['Stayed', 'Left']
attrition_rates
```

```python
import pandas as pd

# Define the current workforce for each generation
workforce_data = {
    "Year": [2024],  # Start from the current year
    "Gen Z": [97],
    "Millennial": [851],
    "Gen X": [453],
    }

# Convert into a DataFrame
df_forecast = pd.DataFrame(workforce_data)

# Display the first row to verify
print(df_forecast)
```

```python
# Define attrition rates per generation
attrition_rates = {
    "Gen Z": 0.3918,  # 39.18% annual attrition
    "Millennial": 0.1669,  # 16.69% annual attrition
    "Gen X": 0.1015  # 10.15% annual attrition
}

# Forecast workforce decline over the next 30 years
years = list(range(2025, 2055))  # 2025 to 2054
for year in years:
    last_row = df_forecast.iloc[-1].copy()  # Get last year's data
    new_row = {"Year": year}

    # Apply attrition to each generation
    for gen in attrition_rates:
        new_row[gen] = last_row[gen] * (1 - attrition_rates[gen])  # Reduce workforce

    # Append the new row to the DataFrame
    df_forecast = pd.concat([df_forecast, pd.DataFrame([new_row])], ignore_index=True)

# Display the first few rows of the forecast
print(df_forecast.head())
```

```python
#Create the visual decline for attrition over 30 years - Linear Regression using Matplotlib

import matplotlib.pyplot as plt

# Create the plot
plt.figure(figsize=(10,5))
plt.plot(df_forecast["Year"], df_forecast["Gen Z"], label="Gen Z", marker="o")
plt.plot(df_forecast["Year"], df_forecast["Millennial"], label="Millennial", marker="o")
plt.plot(df_forecast["Year"], df_forecast["Gen X"], label="Gen X", marker="o")

# Labels and title
plt.xlabel("Year")
plt.ylabel("Remaining Workforce")
plt.title("Projected Workforce Decline Over Time")
plt.legend()
plt.show()
```

Our analysis confirms that Gen Z is leaving the workforce at a rapid pace, with Millennials following closely behind.

Currently, Millennials make up the largest share of the workforce. However, with an average tenure of just six years—compared to Gen X's nine years—this trend suggests a concerning shift. If no changes are made, Gen X, the generation approaching retirement, will become the dominant workforce by 2033.

This poses significant challenges for talent retention, succession planning, and long-term workforce stability.
